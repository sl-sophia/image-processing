```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image_in_name="apple_256x256.png"
image=cv2.imread(image_in_name)
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
image_hsv_h=cv2.cvtColor(image,cv2.COLOR_RGB2HSV)

h=image_hsv_h[:,:,1]>=150

image_hsv_h=cv2.cvtColor(image_hsv_h,cv2.COLOR_HSV2RGB)

image_hsv_h[:,:,0]*=h
image_hsv_h[:,:,1]*=h
image_hsv_h[:,:,2]*=h
```

```python
image_hsv_v=cv2.cvtColor(image,cv2.COLOR_RGB2HSV)

v=image_hsv_v[:,:,2]>150

image_hsv_v=cv2.cvtColor(image_hsv_v,cv2.COLOR_HSV2RGB)

image_hsv_v[:,:,0]*=v
image_hsv_v[:,:,1]*=v
image_hsv_v[:,:,2]*=v
```

```python
plt.figure()

plt.subplot(131)
plt.imshow(image)

plt.subplot(132)
plt.imshow(image_hsv_h)

plt.subplot(133)
plt.imshow(image_hsv_v)

plt.show()
```
