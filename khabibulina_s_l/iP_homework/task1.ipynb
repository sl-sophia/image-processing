```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
import math
```

```python
def gamma(gamma):
    gamma=1/gamma
    return np.array([((i / 255.0) ** gamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
```

```python
def inverse():
    return np.arange(255,-1,-1)
```

```python
def shift():
    return np.append(np.arange(150,256),np.arange(0,150))
```

```python
def three():
    return np.concatenate([np.zeros(85),np.full((85),128),np.full((86),255)])
```

```python
def paint(input_image,func):
    
    plt.figure(figsize=(21,6))
    
    plt.subplot(131)
    plt.imshow(input_image,cmap='gray')
    
    out=cv2.LUT(input_image, func) #обработка
    
    plt.subplot(132)
    plt.imshow(out,cmap='gray')
    
    plt.subplot(133)
    plt.plot(range(0,256),func) #Построение графика            
    plt.xlabel(r'$x$') #Метка по оси x в формате TeX
    plt.ylabel(r'$f(x)$') #Метка по оси y в формате TeX
    plt.title(r'$y=f(x)$') #Заголовок в формате TeX
    plt.grid(True) #Сетка
    
    plt.show() #Показать график
```

```python
image=cv2.imread("s1200.png",cv2.IMREAD_GRAYSCALE) #исходное изображение
```

```python
paint(image,gamma(0.1))
```

```python
paint(image,inverse())
```

```python
paint(image,shift())
```

```python
paint(image,three())
```
