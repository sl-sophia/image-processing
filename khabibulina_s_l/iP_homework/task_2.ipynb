```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_contures(image_name):
    # загрузка изображения, переход в градации серого и уменьшение резкости (фильтрация и уменьшение размера)
    image = cv2.cvtColor(cv2.imread(image_name), cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    #gray = cv2.resize(gray, (100,100), interpolation = cv2.INTER_AREA)
    gray=cv2.medianBlur(gray,17)
    
    # распознавание контуров
    edged = cv2.Canny(gray, 21, 50)
    
    # закрытие контуров
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    closed = cv2.morphologyEx(edged, cv2.MORPH_CLOSE, kernel)
    
    plt.figure(figsize=(21,6))
    
    plt.subplot(131)
    plt.imshow(gray,cmap='gray')
    
    plt.subplot(132)
    plt.imshow(edged,cmap='gray')
    
    plt.subplot(133)
    plt.imshow(closed,cmap='gray')
    
    plt.show()
    
    #контуры в изображении
    cnts = cv2.findContours(closed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[1]
    total=0
    for c in cnts:
    # аппроксимируем (сглаживаем) контур
        #peri = cv2.arcLength(c, True)
        #approx = cv2.approxPolyDP(c, 0.01 * peri, True)
        #cv2.drawContours(image, [approx], -1, (0, 255, 0), 4)
        if len(c)>100:
            #cv2.drawContours(image, [c], -1, (0, 255, 0), 4)
            peri = cv2.arcLength(c, True)
            print(peri)
            approx = cv2.approxPolyDP(c, peri*0.002, True) #0.001
            cv2.drawContours(image, [approx], -1, (0, 255, 0), 3)
    
    return image
```

```python
img=find_contures("seastar1.jpg")
```
